```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
import pandas as pd
import pickle
from sklearn import set_config
set_config(display='diagram')
```

```python
# Load train data
train_data = pd.read_csv('Titanic_train.csv')
X_train = train_data.drop(['Name', 'Pclass', 'Ticket', 'Parch', 'Cabin', 'Survived'], axis=1)  # Replace 'target' with the actual target column name
y_train = train_data['Survived']

# Load test data
test_data = pd.read_csv('Titanic_test.csv')
X_test = test_data.drop(['Parch', 'Name', 'Pclass', 'Ticket', 'Cabin'], axis=1)  # Replace 'target' with the actual target column name (if applicable)
```

```python
# Preprocessing and model
scaler = StandardScaler()
model = LogisticRegression()

# Combine preprocessing and model into a pipeline
pipeline = Pipeline(steps=[
    ('scaling', scaler),
    ('logistic_regression', model)
])

# Fit the pipeline on the training data
pipeline.fit(X_train, y_train)

print("Pipeline training complete.")
```
