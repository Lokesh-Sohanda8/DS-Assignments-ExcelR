# Importance of Data Preprocessing

- Standardization
  1. Standardization is a Critical aspect in Data Preprocessing.
  2. This is the process of converting data into a uniform format ensuring its consistency.
  3. This converts the Mean (average) to 0 & Standard Deviation (spread out of data from Mean) to 1.

- One Hot Encoding
  1. The process of converting the data (Categorical Data) into Binary language that Computers understands.
  2. It converts all the unique values into seperate columns with values 0 (False) & 1 (True).
  3. Each 1 value represents the value present in the Row for other columns & vice versa.
